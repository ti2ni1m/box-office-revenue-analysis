## Data Cleaning

Several columns contain missing values or complex nested data (e.g., JSON-like structures such as genres and cast).

To simplify the analysis:
- Irrelevant or highly incomplete columns were removed
- Only useful numerical and date-related features were retained

```python
train = train.drop(['belongs_to_collection', 'genres', 'crew', 'cast', 'Keywords', 
                  'spoken_languages', 'production_companies', 'production_countries', 'tagline','overview','homepage'], axis=1)
```

## Feature Engineering

The release date was converted into datetime format, and additional features were extracted:
- Release year
- Release month
- Release day
- Release weekday

These features help analyze trends over time.

```python
train['release_date'] = pd.to_datetime(train['release_date'], infer_datetime_format=True)
train['release_day'] = train['release_date'].apply(lambda t: t.day)
train['release_weekday'] = train['release_date'].apply(lambda t: t.weekday())
train['release_month'] = train['release_date'].apply(lambda t: t.month)

train['release_year'] = train['release_date'].apply(lambda t: t.year if t.year < 2018 else t.year -100)
```

## 1. Highest Revenue Movie

We identify the movie with the highest box office revenue.

```python
train[train['revenue'] == train['revenue'].max()]
```

```python
train[['id','title','budget','revenue']].sort_values(['revenue'], ascending=False).head(10).style.background_gradient(subset='revenue', cmap='BuGn')
```

**Insight:**
The highest revenue movie is *The Avengers*, earning over $1.5 billion.  
This highlights the strong financial performance of blockbuster franchise films.

## 2. Most Expensive Movie

I analysed which movie had the highest production budget.

```python
train[train['budget'] == train['budget'].max()]
```

```python
train[['id','title','budget', 'revenue']].sort_values(['budget'], ascending=False).head(10).style.background_gradient(subset=['budget', 'revenue'], cmap='PuBu')
```

**Insight:**
*Pirates of the Caribbean: On Stranger Tides* had the highest budget (~$380M).

## 3. Movie Runtime Analysis

```python
plt.hist(train['runtime'].fillna(0) / 60, bins=40);
plt.title('Distribution of length of film in hours', fontsize=16, color='white');
plt.xlabel('Duration of Movie in Hours')
plt.ylabel('Number of Movies')
```

## 4. Movie Releases by Year

```python
plt.figure(figsize=(20,12))

sns.countplot(
    x=train['release_year'].sort_values(),
    hue=train['release_year'].sort_values(),
    palette="Dark2",
    edgecolor=(0,0,0),
    legend=False
)

plt.title("Movie Release count by Year", fontsize=20)
plt.xlabel('Release Year')
plt.ylabel('Number of Movies Release')
plt.xticks(fontsize=12, rotation=90)
plt.show()
```

```python
train['release_year'].value_counts().head()
```

**Insight:**
Movie releases increased significantly in recent years, peaking around 2013.  
This may reflect growth in the global film industry.

## Release Trends by Month / Day / Weekday

6. In which month most movies are released from 1921 to 2017?

```python
plt.figure(figsize=(20,12))

sns.countplot(
    x=train['release_month'],
    hue=train['release_month'],
    palette="Dark2",
    edgecolor=(0,0,0),
    legend=False
)

plt.title("Movie Release count by Month", fontsize=20)
plt.xlabel('Release Month')
plt.ylabel('Number of Movies Release')
plt.xticks(fontsize=12)
plt.show()
```

```python
train['release_month'].value_counts()
```

7. On which date of month most movies are released?

```python
plt.figure(figsize=(20,12))

sns.countplot(
    x=train['release_day'],
    hue=train['release_day'],
    order=range(1, 32),
    palette="Dark2",
    edgecolor=(0,0,0),
    legend=False
)

plt.title("Movie Release count by Day of Month", fontsize=20)
plt.xlabel('Release Day')
plt.ylabel('Number of Movies Release')
plt.xticks(fontsize=12)
plt.show()
```

```python
train['release_day'].value_counts()
```

8. On which day of week most movies are released?

```python
plt.figure(figsize=(20,12))

sns.countplot(
    x=train['release_weekday'],
    hue=train['release_weekday'],
    order=range(7),
    palette='Dark2',
    legend=False      
)

day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.xlabel('Release Day of Week')
plt.ylabel('Number of Movies Release')
plt.xticks(ticks=range(7), labels=day_labels, fontsize=12)

plt.show()
```

```python
train['release_weekday'].value_counts()
```

**Insight:**
- Most movies are released in September and October  
- Fridays are the most common release day  
- Early days of the month (e.g., 1st) are slightly more frequent
